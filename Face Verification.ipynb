```python
import cv2
import numpy as np
import pandas as pd
import tensorflow 
from tensorflow.keras.models import load_model
import os
```

```python
model = tensorflow.keras.models.load_model("Face_verification_model.hdf5")
```

```python
model.summary()
```

```python
cam = cv2.VideoCapture(0)

facedata =os.getcwd()+"//"+"haarcascade_frontalface_alt.xml"
faceCascade = cv2.CascadeClassifier(facedata)
    
classes = ['Nayeem', 'Sai']

while cam.isOpened():
    
    _, frame = cam.read()
    
    #frame = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = faceCascade.detectMultiScale(gray, scaleFactor=1.1)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2, 5)
        face_crop = frame[y:y + h, x:x + w]
        face_crop = cv2.resize(face_crop, (105, 105))
        face_crop = cv2.cvtColor(face_crop, cv2.COLOR_BGR2GRAY)
        face_crop = face_crop.astype('float32')
        face_crop = np.asarray(face_crop)
        face_crop = face_crop.reshape(-1, face_crop.shape[0], face_crop.shape[1], 1)
        print(face_crop.shape)
        value = model.predict([face_crop, face_crop])
        
        val = np.argmax(value)
        real_value = value[0][val]
        print(real_value)
        #real_values.append(real_value)
        result = classes[val]
        
        print(result)
            
    cv2.imshow("image", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
